# Agregação dos Dados de Não Falha da Aeronave 091
Nesta etapa, realizamos a agregação dos dados referentes à aeronave de identificação 091 que não apresentaram falhas. Esse processo envolveu a reunião e organização cuidadosa das informações relacionadas a ocorrências de falhas nessa aeronave específica. Essa agregação de dados é fundamental para permitir uma análise mais detalhada e específica das falhas que não ocorreram.

### Importação das bibliotecas

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

### Processamento Incremental e Agregação de Dados para a Aeronave 091
Neste bloco de código, apresentamos um processo completo de processamento incremental e agregação de dados para a aeronave de identificação 091. A função read_and_process_parquet lê e processa os arquivos Parquet selecionando colunas específicas e calculando a média agrupada por uma coluna comum. Em seguida, a função process_files_incrementally itera sobre os arquivos Parquet na pasta "airplane_data_091", excluindo arquivos específicos se necessário, e aplica a função de processamento incremental. O conjunto de dados resultante, denominado flights_combined_data_091, contém informações agregadas de acordo com as colunas selecionadas e a coluna comum "phaseOfFlight-1". Isso possibilita uma análise eficiente e abrangente das médias agrupadas em relação às diferentes fases de voo.

```python
def read_and_process_parquet(file_path, selected_columns, common_column):
    df = pd.read_parquet(file_path, columns=selected_columns)
    grouped = df.groupby(common_column).mean()
    return grouped

def process_files_incrementally(folder_path, excluded_files=None, selected_columns=None, common_column=None):
    parquet_files = [file for file in os.listdir(folder_path) if file.endswith('.parquet')]

    if excluded_files:
        parquet_files = [file for file in parquet_files if file not in excluded_files]

    if not parquet_files:
        print("No Parquet files found in the specified folder.")
        return None
    
    dataframes = []
    
    for file in parquet_files:
        file_path = os.path.join(folder_path, file)
        processed_df = read_and_process_parquet(file_path, selected_columns, common_column)
        dataframes.append(processed_df)
    
    combined_df = pd.concat(dataframes)
    
    return combined_df

folder_name = "airplane_data_091"
excluded_files = ['TCRF_ARCHIVE_06120091_20230428193501.parquet', 'TCRF_ARCHIVE_06120091_20230428193502.parquet', 'TCRF_ARCHIVE_06120091_20230515080501.parquet']
filtered_columns = ['phaseOfFlight-1',
                 'message0418DAA-1','message0422DAA-1','amscHprsovDrivF-1a', 'amscHprsovDrivF-1b',
                 'amscHprsovDrivF-2b', 'amscPrsovDrivF-1a',
                 'amscPrsovDrivF-1b', 'amscPrsovDrivF-2b',
                 'basBleedLowPressF-1a', 'basBleedLowPressF-2b',
                 'basBleedLowTempF-1a', 'basBleedLowTempF-2b',
                 'basBleedOverPressF-1a', 'basBleedOverPressF-2b',
                 'basBleedOverTempF-1a', 'basBleedOverTempF-2b',
                 'bleedFavTmCmd-1a', 'bleedFavTmCmd-1b',
                 'bleedFavTmCmd-2a', 'bleedFavTmCmd-2b', 'bleedFavTmFbk-1a',
                 'bleedFavTmFbk-1b', 'bleedFavTmFbk-2b', 'bleedHprsovCmdStatus-1a',
                 'bleedHprsovCmdStatus-1b', 'bleedHprsovCmdStatus-2a',
                 'bleedHprsovCmdStatus-2b', 'bleedHprsovOpPosStatus-1a',
                 'bleedHprsovOpPosStatus-1b', 'bleedHprsovOpPosStatus-2a',
                 'bleedHprsovOpPosStatus-2b', 'bleedMonPress-1a',
                 'bleedMonPress-1b', 'bleedMonPress-2a', 'bleedMonPress-2b',
                 'bleedOnStatus-1a', 'bleedOnStatus-1b', 'bleedOnStatus-2b',
                 'bleedOverpressCas-2a', 'bleedOverpressCas-2b',
                 'bleedPrecoolDiffPress-1a', 'bleedPrecoolDiffPress-1b',
                 'bleedPrecoolDiffPress-2a', 'bleedPrecoolDiffPress-2b',
                 'bleedPrsovClPosStatus-1a', 'bleedPrsovClPosStatus-2a',
                 'bleedPrsovFbk-1a']
common_column = 'phaseOfFlight-1'

flights_combined_data_091 = process_files_incrementally(folder_name, excluded_files, filtered_columns, common_column)
```

### Preenchimento de valores ausentes

```python
for column in flights_combined_data_091.columns:
    nan_start = True
    for i in range(len(flights_combined_data_091)):
        if nan_start and not pd.isna(flights_combined_data_091[column].iloc[i]):
            nan_start = False
            fill_value = flights_combined_data_091[column].iloc[i]
        if nan_start:
            flights_combined_data_091[column].iloc[i] = 0
        else:
            if pd.isna(flights_combined_data_091[column].iloc[i]):
                flights_combined_data_091[column].iloc[i] = fill_value

print(flights_combined_data_091)
```

### Agregação de dados por fase de voo
Neste trecho de código, realizamos a agregação dos dados da aeronave de identificação 091 com base na coluna "phaseOfFlight-1", que representa as diferentes fases de voo. Utilizando a função groupby, agrupamos os dados de acordo com essa coluna e, em seguida, calculamos a média dos valores das colunas restantes para cada fase de voo. O resultado é um novo conjunto de dados denominado flights_combined_data_091, que contém as médias agrupadas por fase de voo. Isso oferece uma visão resumida das tendências e comportamentos médios durante as diversas fases de operação da aeronave.





```python
grouping_column = 'phaseOfFlight-1'

flights_combined_data_091 = flights_combined_data_091.groupby(grouping_column).mean().reset_index()

print(flights_combined_data_091)
```

### Salvando dados agregados em formato Parquet
O conjunto de dados resultante do processo de agregação e cálculo de médias, denominado flights_combined_data_091, é exportado e armazenado no arquivo "flights_combined_data_091.parquet". Isso permite preservar as informações agregadas de maneira eficiente e pronta para análises futuras, simplificando o processo de compartilhamento e reutilização desses dados.

```python
flights_combined_data_091.to_parquet("flights_combined_data_091.parquet")
```
