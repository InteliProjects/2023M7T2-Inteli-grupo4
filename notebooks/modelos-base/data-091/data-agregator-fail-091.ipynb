# Agregação dos Dados de Falha da Aeronave 091
Nesta etapa, realizamos a agregação dos dados referentes à aeronave de identificação 091 que apresentaram falhas. Esse processo envolveu a reunião e organização cuidadosa das informações relacionadas a ocorrências de falhas nessa aeronave específica. ### Importação das bibliotecas

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

### Detectando Variações em Dados da Aeronave 091
Nesta seção, apresentamos um código em Python que tem como objetivo identificar variações nos dados da aeronave de identificação 091. O script utiliza a biblioteca pandas para processar os arquivos no formato Parquet contidos na pasta "airplane_data_091" (pasta apagada que continha todos os arquivos parquet da aeronave 091). As colunas de interesse para análise são "message0418DAA-1" e "message0422DAA-1". O código verifica se há variações significativas em pelo menos uma dessas colunas em relação aos valores 0 e valores ausentes (NaN), que significa que ocorreu falha. Os arquivos que exibem tais variações são registrados e exibidos ao final da execução.


```python
folder_path = "airplane_data_091"

columns_to_check = ['message0418DAA-1', 'message0422DAA-1']

files_with_variation = []

parquet_files = [file for file in os.listdir(folder_path) if file.endswith('.parquet')]

for file in parquet_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_parquet(file_path)
    
    variation_detected = any(df[column].nunique() > 1 for column in columns_to_check if not df[column].isin([0, np.nan]).all())
    
    if variation_detected:
        files_with_variation.append(file)

print("Arquivos com variação nas colunas:", files_with_variation)
```

### Processamento Incremental de Dados para Falhas na Aeronave 091Neste segmento, é apresentado um código em Python que exemplifica o processo de processamento incremental de dados relacionados a falhas na aeronave de identificação 091. A função read_filtered_columns_from_parquet é responsável por ler colunas específicas de um arquivo Parquet. Em seguida, a função process_files_incrementally combina os dados de vários arquivos Parquet, considerando as colunas relevantes previamente filtradas. O código ilustra o uso dessas funções para processar e combinar dados dos arquivos "TCRF_ARCHIVE_06120091_20230428193501.parquet", "TCRF_ARCHIVE_06120091_20230428193502.parquet" e "TCRF_ARCHIVE_06120091_20230515080501.parquet". As colunas selecionadas para processamento incluem informações como o status de diversos componentes, com o objetivo de analisar detalhes das falhas da aeronave de maneira incremental e eficiente.





```python
import os
import pandas as pd

def read_filtered_columns_from_parquet(file_path, filtered_columns):
    df = pd.read_parquet(file_path, columns=filtered_columns)
    return df

def process_files_incrementally(folder_path, selected_files=None,filtered_columns=None):
    parquet_files = [file for file in os.listdir(folder_path) if file.endswith('.parquet')]

    if selected_files:
        parquet_files = [file for file in parquet_files if file in selected_files]

    if not parquet_files:
        print("No Parquet files found in the specified folder.")
        return None
    
    dataframes = []
    
    for file in parquet_files:
        file_path = os.path.join(folder_path, file)
        processed_df = read_filtered_columns_from_parquet(file_path, filtered_columns)
        dataframes.append(processed_df)
    
    combined_df = pd.concat(dataframes)
    
    return combined_df

folder_name = "airplane_data_091"
selected_files = ['TCRF_ARCHIVE_06120091_20230428193501.parquet', 'TCRF_ARCHIVE_06120091_20230428193502.parquet', 'TCRF_ARCHIVE_06120091_20230515080501.parquet']
filtered_columns = ['phaseOfFlight-1',
                 'message0418DAA-1','message0422DAA-1','amscHprsovDrivF-1a', 'amscHprsovDrivF-1b',
                 'amscHprsovDrivF-2b', 'amscPrsovDrivF-1a',
                 'amscPrsovDrivF-1b', 'amscPrsovDrivF-2b',
                 'basBleedLowPressF-1a', 'basBleedLowPressF-2b',
                 'basBleedLowTempF-1a', 'basBleedLowTempF-2b',
                 'basBleedOverPressF-1a', 'basBleedOverPressF-2b',
                 'basBleedOverTempF-1a', 'basBleedOverTempF-2b',
                 'bleedFavTmCmd-1a', 'bleedFavTmCmd-1b',
                 'bleedFavTmCmd-2a', 'bleedFavTmCmd-2b', 'bleedFavTmFbk-1a',
                 'bleedFavTmFbk-1b', 'bleedFavTmFbk-2b', 'bleedHprsovCmdStatus-1a',
                 'bleedHprsovCmdStatus-1b', 'bleedHprsovCmdStatus-2a',
                 'bleedHprsovCmdStatus-2b', 'bleedHprsovOpPosStatus-1a',
                 'bleedHprsovOpPosStatus-1b', 'bleedHprsovOpPosStatus-2a',
                 'bleedHprsovOpPosStatus-2b', 'bleedMonPress-1a',
                 'bleedMonPress-1b', 'bleedMonPress-2a', 'bleedMonPress-2b',
                 'bleedOnStatus-1a', 'bleedOnStatus-1b', 'bleedOnStatus-2b',
                 'bleedOverpressCas-2a', 'bleedOverpressCas-2b',
                 'bleedPrecoolDiffPress-1a', 'bleedPrecoolDiffPress-1b',
                 'bleedPrecoolDiffPress-2a', 'bleedPrecoolDiffPress-2b',
                 'bleedPrsovClPosStatus-1a', 'bleedPrsovClPosStatus-2a',
                 'bleedPrsovFbk-1a']

flights_combined_data_fail_091 = process_files_incrementally(folder_name, selected_files, filtered_columns)
```

### Preenchimento de valores ausentes

```python
for column in flights_combined_data_fail_091.columns:
    nan_start = True
    for i in range(len(flights_combined_data_fail_091)):
        if nan_start and not pd.isna(flights_combined_data_fail_091[column].iloc[i]):
            nan_start = False
            fill_value = flights_combined_data_fail_091[column].iloc[i]
        if nan_start:
            flights_combined_data_fail_091[column].iloc[i] = 0
        else:
            if pd.isna(flights_combined_data_fail_091[column].iloc[i]):
                flights_combined_data_fail_091[column].iloc[i] = fill_value

print(flights_combined_data_fail_091)
```

### Agregação de dados personalizada (média)
A função custom_mean é definida para calcular a média dos valores em uma série, ignorando valores ausentes. Utilizando as colunas de agrupamento definidas em grouping_columns, o conjunto de dados flights_combined_data_fail_091 é agrupado de acordo com essas colunas e, em seguida, a função de agregação personalizada é aplicada a cada grupo. O resultado é uma nova estrutura de dados que contém as médias calculadas para as colunas de interesse, mantendo a estrutura de agrupamento original. 

```python
def custom_mean(series):
    return series.mean(skipna=True)

grouping_columns = ['phaseOfFlight-1',  'message0418DAA-1', 'message0422DAA-1']

flights_combined_data_fail_091 = flights_combined_data_fail_091.groupby(grouping_columns, as_index=False).apply(lambda group: group.apply(custom_mean)).reset_index(drop=True)

print(flights_combined_data_fail_091)
```

### Salvando dados agregados em formato Parquet
O conjunto de dados resultante do processo de agregação e cálculo de médias, denominado flights_combined_data_fail_091, é exportado e armazenado no arquivo "flights_combined_data_fail_091.parquet". Isso permite preservar as informações agregadas de maneira eficiente e pronta para análises futuras, simplificando o processo de compartilhamento e reutilização desses dados.

```python
flights_combined_data_fail_091.to_parquet("flights_combined_data_fail_091.parquet")
```
