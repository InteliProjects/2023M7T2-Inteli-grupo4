```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import warnings
warnings.filterwarnings('ignore')
import math
import shutil
import pyarrow as pa
import pyarrow.parquet as pq
```

```python
teto = 1
```

```python
folder_path = ("airplane_data_035")

parquet_files = [file for file in os.listdir(folder_path) if file.endswith('.parquet')]

flights = len(parquet_files)
print(flights)
range_max = int(math.floor(flights/50))
```

```python
print(range_max)
```

```python
filtered_columns = [ 'phaseOfFlight-1',
                'bleedFavTmCmd-1a', 'bleedFavTmCmd-1b',
                'bleedFavTmCmd-2a', 'bleedFavTmCmd-2b', 'bleedFavTmFbk-1a',
                'bleedFavTmFbk-1b', 'bleedFavTmFbk-2b', 'bleedHprsovCmdStatus-1a',
                'bleedHprsovCmdStatus-1b', 'bleedHprsovCmdStatus-2a',
                'bleedHprsovCmdStatus-2b', 'bleedHprsovOpPosStatus-1a',
                'bleedHprsovOpPosStatus-1b', 'bleedHprsovOpPosStatus-2a',
                'bleedHprsovOpPosStatus-2b', 'bleedMonPress-1a',
                'bleedMonPress-1b', 'bleedMonPress-2a', 'bleedMonPress-2b',
                'bleedOnStatus-1a', 'bleedOnStatus-1b', 'bleedOnStatus-2b',
                'bleedPrecoolDiffPress-1a', 'bleedPrecoolDiffPress-1b',
                'bleedPrecoolDiffPress-2a', 'bleedPrecoolDiffPress-2b',
                'bleedPrsovClPosStatus-1a', 'bleedPrsovClPosStatus-2a',
                'bleedPrsovFbk-1a','message0418DAA-1', 'message0422DAA-1']
```

```python
def convert_columns(df_filtered):
    for coluna in df_filtered.columns:
        if df_filtered[coluna].dtype == 'float64':
            df_filtered[coluna] = df_filtered[coluna].astype('float32')

        if df_filtered[coluna].dtype == 'int64':
            df_filtered[coluna] = df_filtered[coluna].astype('int32')

    return df_filtered
```

```python
df_number = 0
```

```python
def df_consolida(index):
    global df_number
    global flights
    step = 0
    if index < range_max:
        step = teto*index
    else:
        step = flights
    df = pd.DataFrame()
    for _, _, arquivos in os.walk('./airplane_data_035'):
        #nota: para alterar o tamanho do dataset, basta alterar o valor de 250. Recomendo utilizar o valor de 250 para não estourar o tamanho da RAM do colab
        while(df_number != step):
                df_read = pd.read_parquet("./airplane_data_035/" + arquivos[df_number] ,engine='auto')
                df_filtered = df_read[filtered_columns]#
                df_converted = convert_columns(df_filtered)
                df = pd.concat([df, df_converted], axis=0)
                df_number = df_number + 1
                print(df_number)
    return df
    
```

```python
df_consolidado = df_consolida(1)
```

```python
df_consolidado
```

```python
# # saving the dataframe
# df_consolidado.to_csv('file2.csv', index=False)

```

```python
data_fail_agregator_035 = df_consolidado.fillna(method="backfill")
```

```python
df_035_fail_no_nan = data_fail_agregator_035.dropna(axis=0)
df_035_fail_no_nan.isna().sum()
```

```python
df_035_fail_no_nan['message_error'] = df_035_fail_no_nan['message0418DAA-1'] + df_035_fail_no_nan['message0422DAA-1']


def substituir_valor(valor):
    if valor == 2:
        return 1
    else:
        return valor

df_035_fail_no_nan['message_error'].apply(substituir_valor)
print(df_035_fail_no_nan['message_error'])
```

```python
def custom_mean(series):
    return series.mean(skipna=True)

grouping_columns = ['phaseOfFlight-1','message0418DAA-1', 'message0422DAA-1', 'message_error']

flights_combined_data_fail_035 = df_035_fail_no_nan.groupby(grouping_columns, as_index=False).apply(lambda group: group.apply(custom_mean)).reset_index(drop=True)

print(flights_combined_data_fail_035)

```

```python
flights_combined_data_fail_035.to_parquet("flights_combined_data_fail_035.parquet")
df_035_fail_no_nan.to_parquet("agregated_data_035.parquet")
```

```python
def plot_all_columns(dataframe):
    for col in dataframe.columns:
        print(col)
        try:
            plt.figure(figsize=(10, 6))
            plt.plot(dataframe[col], marker='o')
            plt.title(f'Line Plot for Column: {col}')
            plt.xlabel('Index')
            plt.ylabel(col)
            plt.grid()
        except:
            print(f'Could not plot for column: {col}')

plot_all_columns(flights_combined_data_fail_035)
plt.show()

```
